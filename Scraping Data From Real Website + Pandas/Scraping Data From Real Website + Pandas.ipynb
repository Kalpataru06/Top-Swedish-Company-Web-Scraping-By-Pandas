# Scraping Data From Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = r'https://en.wikipedia.org/wiki/List_of_largest_Swedish_companies'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = "wikitable sortable" ,style="text-align:right;")
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
world_titles = soup.find_all('th')[8:17]
```

```python
world_titles
```

```python
world_table_title = [title.text.strip() for title in world_titles ]
```

```python
world_table_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_title )
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
    
    #print(individual_row_data)
```
